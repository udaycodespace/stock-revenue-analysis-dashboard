```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Question 1: Use yfinance to Extract Tesla Stock Data

```python
# Using backup CSV instead of live yfinance for reliability
tesla_data = pd.read_csv("tesla_stock.csv")
display(tesla_data.head(5))
```

## Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
# Using backup CSV instead of live webscraping for reliability
tesla_revenue = pd.read_csv("tesla_revenue.csv")
display(tesla_revenue.tail(5))
```

## Question 3: Use yfinance to Extract GME Stock Data

```python
# Using backup CSV
gme_data = pd.read_csv("gme_stock.csv")
display(gme_data.head(5))
```

## Question 4: Use Webscraping to Extract GME Revenue Data

```python
# Using backup CSV
gme_revenue = pd.read_csv("gme_revenue.csv")
display(gme_revenue.tail(5))
```

```python
def make_graph(stock_data, revenue_data, stock, title):
    fig, ax1 = plt.subplots(figsize=(10,6))
    ax1.plot(stock_data["Date"], stock_data["Close"], label="Stock Price", color="blue")
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Stock Price (USD)", color="blue")
    ax2 = ax1.twinx()
    ax2.plot(revenue_data["Date"], revenue_data["Revenue"], label="Revenue", color="red")
    ax2.set_ylabel("Revenue (USD)", color="red")
    plt.title(title)
    fig.tight_layout()
    plt.show()
```

## Question 5: Plot Tesla Stock Graph

```python
make_graph(tesla_data, tesla_revenue, "Tesla", "Tesla Stock Price vs Revenue")
```

## Question 6: Plot GME Stock Graph

```python
make_graph(gme_data, gme_revenue, "GameStop", "GameStop Stock Price vs Revenue")
```
